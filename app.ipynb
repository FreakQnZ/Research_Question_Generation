```python
import pandas as pd 
from PyPDF2 import PdfReader as pdf
from matplotlib import pyplot as plt
import io
from PIL import Image
import easyocr
```

```python
reader = pdf("uploads/mech.pdf")
```

```python
text = reader.pages[2].extract_text()
print(text)
```

```python
image = reader.pages[2].images[0]

with open ("image.png", "wb") as f:
    f.write(image.data)

image_data = io.BytesIO(image.data)
img = Image.open(image_data)

plt.imshow(img)
plt.axis("off")
plt.show()
```
