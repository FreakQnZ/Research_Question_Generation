```python
import pandas as pd 
import numpy as np
from PyPDF2 import PdfReader as pdf
from matplotlib import pyplot as plt
import io
from PIL import Image
import easyocr
import spacy
```

```python
reader = pdf("uploads/evs.pdf")
```

## Text Preprocessing

```python
def read_Pages(pages):
    text = []
    remove_chars = ["•", "❑", "➢"]
    for page in pages:
        content = page.extract_text()
        content = content.strip()
        for char in remove_chars:
            content = content.replace(char, "")
        text.append(content)
    text = " ".join(text)
    return text
```

```python
text = read_Pages(reader.pages)
print(text)
```

```python
# Writing text to txt file

with open ("uploads/evs.txt", "w") as f:
    f.write(text)
```
